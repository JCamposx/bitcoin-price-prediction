```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('tweets.csv', sep=';', on_bad_lines='skip',
                   lineterminator='\n', low_memory=False)
data
```

```python
data = data[['timestamp', 'text\r']]
data = data.rename(columns={'timestamp': 'date', 'text\r': 'text'})
data
```

```python
data.isna().sum()
```

```python
from datetime import datetime

data.date = np.array(
    [datetime.strptime(d.split(' ')[0], '%Y-%m-%d') for d in data.date])
```

```python
data.sort_values(by='date', inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python
def check_skipped_days():
    x = 0
    arr = np.array([])
    day_before = datetime.now()
    index_before = 0

    for index, d in enumerate(data.date):
        diff = 0

        if x == 0:
            day_before, index_before = d, index
            x+=1
            continue

        diff = (d - day_before).days

        if diff > 1:
            arr = np.append(arr, '{} => {} : {} days'.format(index_before, index, diff))

        day_before, index_before = d, index

    return arr
```

```python
skipped_days = check_skipped_days()

print(skipped_days)
len(skipped_days)
```

```python
data = data.loc[207:]
data.reset_index(drop=True, inplace=True)
```
